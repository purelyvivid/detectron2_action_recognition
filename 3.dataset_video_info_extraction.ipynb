```python
import pandas as pd
```

```python
%%time
video_info = []
for fn in filenames:
    _, video_dict = load_video(fn)
    video_info.append([fn.split('/')[-1], fn.split('/')[-2], \
                       video_dict['width'], video_dict['height'],
                       video_dict['frames_per_second'], video_dict['num_frames'],
                      ])
video_info_df = pd.DataFrame(video_info, columns=['filename', 'class', 'width', 
                                                  'height', 'frames_per_second', 'num_frames'])
video_info_df.shape
```

```python
video_info_df.to_csv("demo/video_info.csv")
```
